```python
import json
import os
import cv2

def via_to_coco(via_json_path, image_dir, output_json_path):
    with open(via_json_path) as f:
        via_data = json.load(f)

    coco_output = {
        "info": {
            "description": "Building Dataset",
            "version": "1.0",
            "year": 2025,
            "contributor": "Rubel",
            "date_created": "2025-08-05"
        },
        "licenses": [
            {
                "id": 1,
                "name": "CC-BY-4.0",
                "url": "https://creativecommons.org/licenses/by/4.0/"
            }
        ],
        "images": [],
        "annotations": [],
        "categories": [
            {"id": 1, "name": "building", "supercategory": "structure"}
        ]
    }

    annotation_id = 1
    for img_id, (filename, entry) in enumerate(via_data.items()):
        image_path = os.path.join(image_dir, entry["filename"])
        image = cv2.imread(image_path)
        if image is None:
            print(f"Image not found: {image_path}")
            continue
        height, width = image.shape[:2]

        coco_output["images"].append({
            "id": img_id,
            "file_name": entry["filename"],
            "width": width,
            "height": height
        })

        # ✅ Handle regions being either dict or list
        if isinstance(entry['regions'], dict):
            regions = list(entry['regions'].values())
        else:
            regions = entry['regions']

        for region in regions:
            shape = region["shape_attributes"]
            if shape["name"] != "polygon":
                continue

            all_points_x = shape["all_points_x"]
            all_points_y = shape["all_points_y"]
            segmentation = [[x for pair in zip(all_points_x, all_points_y) for x in pair]]

            x_min, y_min = min(all_points_x), min(all_points_y)
            x_max, y_max = max(all_points_x), max(all_points_y)
            bbox = [x_min, y_min, x_max - x_min, y_max - y_min]

            coco_output["annotations"].append({
                "id": annotation_id,
                "image_id": img_id,
                "category_id": 1,
                "segmentation": segmentation,
                "bbox": bbox,
                "area": (x_max - x_min) * (y_max - y_min),
                "iscrowd": 0
            })
            annotation_id += 1

    # Save to JSON
    with open(output_json_path, "w") as f:
        json.dump(coco_output, f, indent=4)

    print(f"COCO format JSON saved to: {output_json_path}")
```

```python
via_to_coco(
    via_json_path="Spacenet/train/via_region_data.json",
    image_dir="Spacenet/train/",
    output_json_path="Spacenet/train/train_coco_format.json"
)
```

```python
from detectron2.data.datasets import register_coco_instances

register_coco_instances("train_coco1", {}, "Spacenet/train/train_coco_format.json", "Spacenet/train/")
register_coco_instances("val_coco1", {}, "Spacenet/val/val_coco_format.json", "Spacenet/val/")
```

```python
from detectron2.config import get_cfg
from detectron2 import model_zoo

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_1x.yaml"))

cfg.DATASETS.TRAIN = ("train_coco1",)
cfg.DATASETS.TEST = ("val_coco1",)

cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1  # Only 'building'
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_1x.yaml")  # Pretrained weights

cfg.SOLVER.IMS_PER_BATCH = 4  # or 2 if GPU memory is low
cfg.SOLVER.BASE_LR = 0.0005  # slightly lower to avoid overfitting
cfg.SOLVER.MAX_ITER = 5000   # increase for better learning
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128  # increase learning capacity

cfg.OUTPUT_DIR = "Spacenet/output_retrained"
cfg.INPUT.RANDOM_FLIP = "none" # good for augmentation
```

```python
from detectron2.data import MetadataCatalog

MetadataCatalog.get("train_coco1").thing_classes = ["building"]
MetadataCatalog.get("val_coco1").thing_classes = ["building"]
```

```python
from detectron2.engine import DefaultTrainer
import os

os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

```python
import json, os, cv2
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import geopandas as gpd

from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer
from detectron2 import model_zoo
from detectron2.data import MetadataCatalog

def plot_prediction_and_ground_truth(image_path, json_path, model_weights):
    image_name = os.path.basename(image_path)

    # ---- Setup model config ----
    cfg = get_cfg()
    cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_1x.yaml"))
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
    cfg.MODEL.WEIGHTS = model_weights
    cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1
    cfg.MODEL.DEVICE = "cpu"
    predictor = DefaultPredictor(cfg)

    # Get metadata
    metadata = MetadataCatalog.get(cfg.DATASETS.TRAIN[0])


    # ---- Read image and predict ----
    im = cv2.imread(image_path)
    outputs = predictor(im)

    v = Visualizer(im[:, :, ::-1], metadata=metadata, scale=1.0)
    out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
    predicted_img = out.get_image()[:, :, ::-1]

    plt.imshow(predicted_img)



image_path = "Spacenet/test/Screenshot 2025-08-05 142350.png"
json_path = "Spacenet/val/val_coco_format.json"
model_weights = "Spacenet/output_retrained/model_final.pth"
plot_prediction_and_ground_truth(image_path, json_path, model_weights)
```
